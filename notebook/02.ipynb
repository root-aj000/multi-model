```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the design style for charts
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6) # Default plot size

print("Libraries loaded successfully.")
```

```python
# Load the dataset
df = pd.read_csv('U:\project\_dataset_gen\main_cleaned.csv')

# Show the total count and the first few rows
print(f"Total Rows: {len(df)}")
df.head()
```

```python
# Find rows where text is repeated
# keep=False means we show ALL instances (original + copy)
duplicates = df[df.duplicated(subset=['text'], keep=False)].sort_values(by='text')

if len(duplicates) > 0:
    print(f"⚠️ Found {len(duplicates)} rows with duplicate text!")
    # Display the specific columns to check
    display(duplicates[[ 'text']])
else:
    print("✅ No text duplicates found.")
```

```python
plt.figure(figsize=(12, 8))

# 1. Extended Color Map (includes the missing colors from your error)
base_color_map = {
    'Red': '#d62728', 'Blue': '#1f77b4', 'Green': '#2ca02c', 
    'Yellow': '#ff7f0e', 'Black': '#222222', 'White': '#f0f0f0',
    'Orange': '#ff9f1c', 'Purple': '#9b5de5', 'Pink': '#f15bb5', 
    'Grey': '#808080', 'Brown': '#8B4513', 'Tan': '#D2B48C', 
    'Turquoise': '#40E0D0', 'Navy': '#000080', 'Beige': '#F5F5DC', 
    'Cream': '#FFFDD0', 'Champagne': '#F7E7CE', 'Gold': '#FFD700', 
    'Silver': '#C0C0C0', 'Amber': '#FFBF00', 'Cyan': '#00FFFF'
}

# 2. Safety Step: Create a palette that covers EVERY value in your data
# If a value isn't in the map above, it gets assigned a default grey (#cccccc)
unique_vals = df['dominant_colour'].unique()
safe_palette = {}

for val in unique_vals:
    # .get() looks for the color, if not found returns '#cccccc' (light grey)
    safe_palette[val] = base_color_map.get(val, '#cccccc')

# 3. Order bars by frequency
order = df['dominant_colour'].value_counts().index

# 4. Plot
sns.countplot(
    data=df, 
    y='dominant_colour', 
    order=order, 
    palette=safe_palette, 
    hue='dominant_colour', 
    legend=False,
    edgecolor='black' # Adds a thin border to make light colors visible
)

plt.title('Dominant Background Colors Used', fontsize=15)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Chart 1: Top 10 Target Audiences
top_audiences = df['target_audience'].value_counts().head(10).index
sns.countplot(data=df[df['target_audience'].isin(top_audiences)], 
              y='target_audience', ax=axes[0], palette='mako')
axes[0].set_title('Top Target Audiences', fontsize=14)

# Chart 2: Does Sentiment affect Predicted CTR?
sns.countplot(data=df, x='predicted_ctr', hue='sentiment', ax=axes[1], palette='coolwarm')
axes[1].set_title('Predicted CTR vs Sentiment', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# Remove duplicates, keeping only the first occurrence
df_clean = df.drop_duplicates(subset=['text'], keep='first')

print(f"Original size: {len(df)}")
print(f"Cleaned size:  {len(df_clean)}")

# Save to new file
df_clean.to_csv('main_cleaned.csv', index=False)
print("✅ Saved cleaned data to 'main_cleaned.csv'")
```
